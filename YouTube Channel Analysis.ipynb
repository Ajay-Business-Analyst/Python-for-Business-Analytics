- ### 3.5: Display the names of the top 5 categories based on total views, with minimum 10,000 comments on each video. (4 marks)

```python
# Write your code here to select top 5 category ids with minimum 10000 comments in each video (2 marks)

# Filtering comments count greater than and equal to 10000
# Group by Categories id and sum all numeric values
# Filter top 5 Categories based on highest number of total views
Top5 = videos[videos["comment_count"] >= 10000].groupby("category_id").sum().nlargest(5, "views")

# Capture the Top 5 categories in list
# Using index method to return the first occurrence of the value
ids = list(Top5.index)
ids

```

```python
# Reading Categories csv file
# Categories=pd.read_csv("Categories.csv",index_col="id")

Categories=pd.read_csv("Categories.csv")
Categories
```

```python
#create a function that takes in a list of ids and outputs a list of category names coresponding to those ids (2 marks)
def id_to_name(id_list):
    
    # List to hold Categories values 
    Top5 = []
    
    # Iterating through Category id using for loop
    for i in id_list:
        
        # Updating the list with Category values 
        Top5.append(Categories[Categories["id"]==i]["Category"].values[0])
    
    # Returing the values to the list
    return Top5

# Printing the ids list value which is derived from above above step to get top 5 categories
print(id_to_name(ids))
```

## Task 4: Channel Analysis




- ### 4.1: Find the most popular channel:
    - by the highest total likes (1 mark)
    - by the highest total views (1 mark)




```python
# Grouping videos by Channel title and summing the Numeric data
channel1 = videos.groupby("channel_title").sum()

# Displaying highest total likes from the most popular channel by Filtering based on the highest number of likes
Most_popular_channel_with_Highest_Total_Likes = channel1.nlargest(1, "likes")
print("_____________________________________________________________")
print("Most popular channel with the highest number of total Likes :")
print("_____________________________________________________________")
Most_popular_channel_with_Highest_Total_Likes
```

```python
# Grouping videos by Channel title and summing the Numeric data
channel1 = videos.groupby("channel_title").sum()

# Displaying highest total views from the most popular channel by Filtering based on the highest number of views
Most_popular_channel_with_Highest_Total_Views = channel1.nlargest(1, "views")
print("_____________________________________________________________")
print("Most popular channel with the highest number of total Views :")
print("_____________________________________________________________")
Most_popular_channel_with_Highest_Total_Views
```


- ### 4.2: For the two selected channels, check which one has a higher number of negative responses per positive response. To do this, follow these steps:
